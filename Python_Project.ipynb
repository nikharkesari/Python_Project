```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
stock_url = 'https://github.com/nikharkesari/Python_Project/raw/refs/heads/main/Data_on_assets/S&P500_stock_price_index.xlsx'
bond_url = 'https://github.com/nikharkesari/Python_Project/raw/refs/heads/main/Data_on_assets/US_Aggregate_bond_price_index.xlsx'
rf_url = 'https://github.com/nikharkesari/Python_Project/raw/refs/heads/main/Data_on_assets/Risk-free_rate_of_return.xlsx'
```

```python
stock_data = pd.read_excel(stock_url, parse_dates=True, index_col=0)
bond_data = pd.read_excel(bond_url, parse_dates=True, index_col=0)
rf_data = pd.read_excel(rf_url, parse_dates=True, index_col=0)
```

```python
#calculating excess returns
stock_return = stock_data.pct_change().dropna()
bond_return = bond_data.pct_change().dropna()

start_date = '1980-01-01'
end_date = '2023-12-31'

stock_return = stock_return[start_date:end_date]
bond_return = bond_return[start_date:end_date]
rf_data = rf_data[start_date:end_date]

stock_excess = stock_return.iloc[:, 0] - rf_data.iloc[:, 0]
bond_excess = bond_return.iloc[:, 0] - rf_data.iloc[:, 0]

excess_returns = pd.DataFrame({
    'Stock_Excess': stock_excess,
    'Bond_Excess': bond_excess
})

print("Excess Returns\n")
excess_returns.head()
```

```python
#calculating stats
stats_df = pd.DataFrame(index=excess_returns.columns)
stats_df['Annualized Mean'] = excess_returns.mean() * 12
stats_df['Annualized Volatility'] = excess_returns.std() * np.sqrt(12)
stats_df['Annualized Sharpe'] = stats_df['Annualized Mean'] / stats_df['Annualized Volatility']
stats_df['Skewness'] = excess_returns.apply(lambda x: stats.skew(x))
stats_df['Kurtosis'] = excess_returns.apply(lambda x: stats.kurtosis(x))
stats_df
```
